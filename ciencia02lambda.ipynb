Declaramos un conjunto de numeros.... Lambda se usa para conjuntos!!! es como el for de los programadores

```python
listaNumeros=[4, 7, 9, 33, 88, 74, 57]
listaNumeros
```

lambda nos permite ejecturar tareas completas pero solamente con una linea, por ejemplo:
quiero multimplicar cada numero de la lista por 2 y ver su doble
Para recorrer cada elemento en lambda, tenemos map

```python
# listaNumdoble=[]
# for num in listaNumeros:
#      doble=num*2
#      listaNumDoble.append(doble)
# listaNumDoble
listaNumDoble=list(map(lambda x:x*2, listaNumeros))
listaNumDoble
```

Otra utilidad radica en realizar filtros sobre un conjunto, por ejemplo recurperar numeros positivos 

```python
lista=[2,4,-5,-788,22,1,-99]
listaPositivos=list(filter(lambda x:x>0, lista))
listaPositivos
```
